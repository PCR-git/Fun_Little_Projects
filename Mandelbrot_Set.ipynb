```python
# Imports

import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams

import random
import math
from scipy.linalg import eig

plt.rcParams['figure.figsize'] = [10, 10]
```

```python
# z**2 = (a + ib)^2 = a^2 - b^2 + 2i ab
def f_complex_square(a,b):
  a2 = a**2 - b**2
  b2 = 2*a*b
  return a2, b2

def Quad_map(c1,c2,it=1000):
  a = c1
  b = c2
  for i in np.arange(it):
    a, b = f_complex_square(a,b)
    a += c1
    b += c2
    if np.abs(a) > 1e10 or np.abs(b) > 1e10:
      break
  return(i)
```

```python
div = Quad_map(0,0,it=1000)
print(div)
```

```python
def f_Mandelbrot(res,x_L=2.5,y_L=2,x_i=-1.75,y_i=-1,it=50):
  x_l = int(x_L*res)
  y_l = int(y_L*res)
  x_v = np.arange(x_l+1)/res + x_i
  y_v = np.flip(np.arange(y_l+1)/res + y_i)

  M = np.zeros((y_l,x_l))
  for i in np.arange(x_l):
    for j in np.arange(y_l):
      x = x_v[i]
      y = y_v[j]
      M[j,i] = Quad_map(x,y,it=it)

  return (M * 255/999)
```

```python
res = 400
it = 50
M = f_Mandelbrot(res=res,it=it)
plt.imshow(M)
```
